```python
def paid_less(paid,price):
    #surma bill clear na garna la pakkai ni pachi tip didaina
    needed=price-paid
    print(f"You Need Rs.{needed} more to clear your bill\n")
    try:
        now=int(input("Please! pay the remanining cash:"))
        if type(now) != type(1):
            print("Enter money in digits")
            raise
        if now>needed:
            print("Please enter the needed amount only not more than that")
            raise
    except:
        print("Inside excpet")
        
        paid_less(paid,price)
        
    else:
        needed=needed-now
        if needed==0:
            print("THANK YOU FOR YOUR SERVICE")
            return True
        else:
            paid=paid+now
            paid_less(paid,price)

        
    
    
  
```

```python
def fill_room(roomType,available,name,pn):
    import sqlite3
    if roomType==1:
        for i in available:
            print(i,end=" ")
        while True:    
            choice=int(input("CHOOSE BETWEEN THE FOLLOWING ROOMS:"))
            try:
                if choice not in available:
                    raise
            except:
                print("\tPLEASE!!!\nENTER THE VALID CHOICES ONLY")
                continue
            else:    
                conn = sqlite3.connect("StdRoom.db")
                cur=conn.cursor()
                cur.execute("UPDATE StdRoom SET Name=? ,PhoneNo=? ,Filled=1 WHERE RoomNo=?",(name,pn,choice))
                conn.commit()
                conn.close()
                break
    elif roomType==2:
        for i in available:
            print(i,end=" ")
        while True:    
            choice=int(input("CHOOSE BETWEEN THE FOLLOWING ROOMS:"))
            try:
                if choice not in available:
                    raise
            except:
                print("\tPLEASE!!!\nENTER THE VALID CHOICES ONLY")
                continue
            else:   
                conn = sqlite3.connect("DelRoom.db")
                cur = conn.cursor()
                cur.execute("UPDATE DelRoom SET Name=? ,PhoneNo=? ,Filled=1 WHERE RoomNo=?",(name,pn,choice))
                conn.commit()
                conn.close()
                break
                
                
    elif roomType==3:
        for i in available:
            print(i,end=" ")
        while True:    
            choice=int(input("CHOOSE BETWEEN THE FOLLOWING ROOMS:"))
            try:
                if choice not in available:
                    raise
            except:
                print("\tPLEASE!!!\nENTER THE VALID CHOICES ONLY")
                continue
            else:  
                conn = sqlite3.connect("SDelRoom.db")
                cur=conn.cursor()
                cur.execute("UPDATE SDelRoom SET Name=? ,PhoneNo=? ,Filled=1 WHERE RoomNo=?",(name,pn,choice))
                conn.commit()
                conn.close()
                break
```

```python

def stdRoom():
    conn = sqlite3.connect("StdRoom.db")
    cur = conn.cursor()
    n=[i for i, in cur.execute("SELECT RoomNo FROM StdRoom  WHERE Filled=0")]
    if(n != []):
        print("STANDARD ROOM IS AVAILABLE")
        name,pn =registration()
        fill_room(1,n,name,pn)
    else:
        print("STANDARD ROOM IS FULL")
    conn.close()    
        
def delRoom():
    conn = sqlite3.connect("DelRoom.db")
    cur = conn.cursor()
    n=[i for i, in cur.execute("SELECT RoomNo FROM DelRoom  WHERE Filled=0")]
    #available room ko list nikalxa one line looping use gara ra
    if(n!= []):
        print("DELUXE ROOM IS AVAILABLE")
        name,pn =registration()
        fill_room(2,n,name,pn)
    else:
        print("DELUXE ROOM IS FULL")
    conn.close()
    
    
def sDelRoom():
    conn = sqlite3.connect("SDelRoom.db")
    cur = conn.cursor()
    n=[i for i, in cur.execute("SELECT RoomNo FROM SDelRoom  WHERE Filled=0")]
    if(n!= []):
        print("SUPER DELUXE ROOM IS AVAILABLE")
        name,pn =registration()
        fill_room(3,n,name,pn)
    else:
        print("SUPER DELUXE ROOM IS FULL")
    conn.close()    
                
    
```

```python
def billing(name,pd):
    global price
    #price=0#comment this
    z=0
    print("\n"*10)
    for x in bill.values():
        z=z+x
    print("Total Bill={}".format(z))
    
    paid=int(input("Enter the amount of money paid by the customer:"))
    if paid<z:
        paid_less(paid,z)
        print("\n"*5)
        print(f"BILL FOR {name}    PHONE NUMBER {pd}")
        print("_"*41)
        for x,y in bill.items():
            print("|"+" "+x+"---------->"+str(y)+"\t\t|")
            price=price+y
        print("|"+"-"*39+"|")
        print("|"+"TOTAL"+ "-"*21+">"+"Rs."+str(price)+"\t"+"|")
        print("|"+"-"*39+"|")
        print("*"*19+"PAID"+"*"*18)
        
    else:
        print("\n"*5)
        print(f"BILL FOR {name}    PHONE NUMBER {pd}")
        print("_"*41)
        for x,y in bill.items():
            print("|"+" "+x+"---------->"+str(y)+"\t\t|")
            price=price+y
        print("|"+"-"*39+"|")
        print("|"+"TOTAL"+ "-"*19+">"+"Rs."+str(price)+"\t"+"|")
        print("|"+"PAID"+"-"*20+">"+"Rs."+str(paid)+"\t"+"|")
        print("|"+"RETURN"+"-"*18+">"+"Rs."+str(paid_more(paid,price))+"\t\t"+"|")
        print("|"+"-"*39+"|")
        print("*"*19+"PAID"+"*"*18)

    

def paid_more(paid,price):
    ret=paid- price
    return ret
    

def paid_less(paid,price):
    #surma bill clear na garna la pakkai ni pachi tip didaina
    needed=price-paid
    print(f"You Need Rs.{needed} more to clear your bill\n")
    try:
        now=int(input("Please! pay the remanining cash:"))
        if type(now) != type(1):
            print("Enter money in digits")
            raise
        if now>needed:
            print("Please enter the needed amount only not more than that")
            raise
    except:
        paid_less(paid,price)
        
    else:
        needed=needed-now
        if needed==0:
            print("THANK YOU FOR YOUR SERVICE")
            return True
        else:
            paid=paid+now
            paid_less(paid,price)
```
